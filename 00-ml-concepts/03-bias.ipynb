This notebook demonstrate the impact of intercept(bias) on a linear equation.

And refer to this [article](https://www.geeksforgeeks.org/effect-of-bias-in-neural-network/) for the effect of bias on Neural Network.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Example 1.

Demonstrate the impact of slope (weight) in a linear equation.

```python
# 1st linear equation Definition

# Define the slope and intercept
slope1 = 2
intercept1 = 0

# Generate x values
x1 = np.linspace(-5, 5, 100)

# Calculate y values using the slope-intercept form
y1 = slope1 * x1 + intercept1
```

```python
# 2nd linear equation Definition

# Define the slope and intercept
slope2 = 8
intercept2 = 0

# Generate x values
x2 = np.linspace(-5, 5, 100)

# Calculate y values using the slope-intercept form
y2 = slope2 * x2 + intercept2
```

```python
# Plot the line
fig, axe = plt.subplots()

axe.plot(x1, y1, color='blue')
axe.plot(x2, y2, color='green')

# Emphasize zero on both axes
axe.spines['left'].set_position('zero')
axe.spines['bottom'].set_position('zero')
axe.spines['right'].set_color('none')
axe.spines['top'].set_color('none')

plt.text(x=5, y=15, s=f"linear equation 1: slope value at {slope1}", fontsize=12, color='blue')
plt.text(x=5, y=45, s=f"linear equation 2: slope value at {slope2}", fontsize=12, color='green')



# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

Example 2.

Demonstrate the impact of intercept (bias) in a linear equation.

```python
# 1st linear equation Definition

# Define the slope and intercept
slope1 = 2
intercept1 = 0

# Generate x values
x1 = np.linspace(-10, 10, 100)

# Calculate y values using the slope-intercept form
y1 = slope1 * x1 + intercept1
```

```python
# 2nd linear equation Definition

# Define the slope and intercept
slope2 = 2
intercept2 = -10

# Generate x values
x2 = np.linspace(-10, 10, 100)

# Calculate y values using the slope-intercept form
y2 = slope2 * x2 + intercept2
```

```python
# Plot the line
fig, axe = plt.subplots()

axe.plot(x1, y1, color='blue')
axe.plot(x2, y2, color='green')

# Emphasize zero on both axes
axe.spines['left'].set_position('zero')
axe.spines['bottom'].set_position('zero')
axe.spines['right'].set_color('none')
axe.spines['top'].set_color('none')

plt.text(
    x=12, 
    y=20, 
    s=f"linear equation 1: intercept(bias) value at {intercept1}", 
    fontsize=12, 
    color='blue')

plt.text(
    x=12, 
    y=10, 
    s=f"linear equation 2: intercept(bias) value at {intercept2}", 
    fontsize=12, 
color='green')



# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
